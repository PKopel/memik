```python
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister
```

```python
def inc(circ: QuantumCircuit) -> QuantumCircuit:
    circ.ccx(2,1,0)
    circ.cx(2,1)
    return circ
```

```python
def dec(circ: QuantumCircuit) -> QuantumCircuit:
    circ.x(2)
    circ.x(1)
    circ.ccx(2,1,0)
    circ.x(1)
    circ.cx(2,1)
    circ.x(2)
    return circ
```

```python

circ = QuantumCircuit(3,2)

circ.h(2)

circ = inc(circ)
circ = dec(circ)

circ.h(2)

circ = inc(circ)
circ = dec(circ)

circ.measure(0,0)
circ.measure(1,1)


# circ.qasm(formatted=True, filename='teleportation.qasm')

circ.draw('mpl')
```
